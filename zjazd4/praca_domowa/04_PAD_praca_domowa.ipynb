Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;

```python
dfn = df["tshirt_category"].copy()
```

```python
dfn = dfn.apply(lambda x: x.replace("Wh Tshirt M", "White T-Shirt M"))
dfn = dfn.apply(lambda x: x.replace("Bl Tshirt M", "Black T-Shirt M"))
dfn = dfn.apply(lambda x: x.replace("Wh Tshirt F", "White T-Shirt F"))
dfn = dfn.apply(lambda x: x.replace("Bl Tshirt F", "Black T-Shirt F"))
df["tshirt_category"] = dfn
df["tshirt_category"].value_counts()
```

Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.

```python
def extract_tshirt_type(tc: str) -> str:
  if tc == "Hoodie":
    return "Hoodie"
  elif tc == "Tennis Shirt":
    return "Other"
  else:
    return "T-shirt"
```

```python
def extract_tshirt_size(tc: str) -> str:
  value = tc.split()
  if len(value) == 3:
    return value[2]
  else:
    return None
```

```python
def extract_tshirt_colour(tc: str) -> str:
  value = tc.split()
  if len(value) == 3:
    return value[0]
  else:
    return None
```

```python
df["tshirt_type"] = df['tshirt_category'].apply(lambda x: extract_tshirt_type(x))
df["tshirt_size"] = df['tshirt_category'].apply(lambda x: extract_tshirt_size(x)) 
df["tshirt_colour"] = df['tshirt_category'].apply(lambda x: extract_tshirt_colour(x))
df.head(22)
```

Znajdź zamówienia złożone między 31 grudnia 2014 a 2 sierpnia 2016.

```python
from_date = "2014-12-31"
to_date   = "2016-08-02"
orders = df.query("order_date > @from_date and order_date < @to_date")
orders
```
